#### read in updated dataframe


```python
all_data=pd.read_csv('all_data.csv')
all_data.head()
```

#### data cleaning

##### drop rows having NaN values

```python
nan_df=all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

```python
all_data=all_data.dropna(how='all') #drops rows where all of the values are NaN 
all_data.head()
all_data = all_data [all_data['Order Date'].str[0:2]!='Or']
```

 #### Augment data with additional columns


#### adding month column

```python
all_data['month'] = all_data['Order Date'].str[0:2] #taking string value of date and representing it as month
all_data['month'] = all_data['month'].astype ('int32')
all_data.head()
```

#### adding sales column

```python
# we have quantity and price for each from which we can get the sales column - by multiplying them with each other
#but multiplying directly gives the error " can't multiply sequence by non-int of type 'str'"
#therefore have to convert columns to the correct type
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
all_data['Sales']=all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Q1: best month for sales and how much was earned that month

```python
all_data.groupby('month').sum() 
```

#### december (12) was the best month for sales with total quantity being 28,114 

```python
import matplotlib.pyplot as plt 
months=range(1,13)
results=all_data.groupby('month').sum() 
plt.bar(months,results['Sales'])
plt.xticks( months)
plt.ylabel('Sales in Million USD($)')
plt.xlabel('Month Number')
plt.show()
```

 #### Q2: City with highest sales

#have to add a 'City' column. 
#### adding City column

```python
#extracting city name from the purchase address using .apply method  
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
all_data['City']= all_data['Purchase Address'].apply (lambda x: get_city(x)+' ' + get_state(x))
all_data.head()
```

```python
results=all_data.groupby('City').sum()
results
```

```python
cities = [city for city, df in all_data.groupby('City')]
results=all_data.groupby('City').sum() 
plt.bar(cities,results['Sales'])
plt.xticks( cities,rotation='vertical')
plt.ylabel('Sales in Million USD($)')
plt.xlabel('City')
plt.show()
```

####  Q3: What time should the ads be displayed to maximize the customers buying product

```python
all_data['Order Date']= pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour']=all_data['Order Date'].dt.hour
all_data['Minute']=all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [Hour for Hour, df in all_data.groupby('Hour')]
plt.plot(hours,all_data.groupby(['Hour']).count() ) 
plt.xticks(hours)
plt.xlabel('hours')
plt.ylabel('number of hours')
plt.grid()
plt.show()
```

#### for max sales, the ads should be either at 11am or 7pm(19th hour)

#### Q4: what products are often sold together

```python
# how to solve?
# same order ID => items bought together
```

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)] #gives entries having same order IDs
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
#transform function- executes function for each value
df=df[['Order ID','Grouped']].drop_duplicates() #dropping duplicates 
df.head()
```

```python
#counting most-occuring pairs
from itertools import combinations
from collections import Counter
count=Counter() 
for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
count.most_common (10)
```

#### most commonly bought pair is iPhone and lightning charging cable (1005 times)

#### Q5: what product was sold the most


```python
#  need sum of the quantity ordered, then group them
product_group=all_data.groupby('Product')
quantity_ordered=product_group.sum()['Quantity Ordered']
products = [product for product, df in product_group]
plt.bar(products,quantity_ordered)
plt.xticks( products,rotation='vertical')
plt.ylabel('Quantity')
plt.xlabel('Product')
plt.show()
```
